```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
np.random.seed(0)
cluster1 = np.random.normal(loc=[0, 0], scale=0.5, size=(100, 2))
cluster2 = np.random.normal(loc=[2, 2], scale=0.5, size=(100, 2))
cluster3 = np.random.normal(loc=[-2, 2], scale=0.5, size=(100, 2))
noise = np.random.uniform(low=-4, high=4, size=(50, 2))
data = np.vstack([cluster1, cluster2, cluster3, noise])
```

```python
plt.scatter(data[:, 0], data[:, 1], s=10)
plt.title("Dataset with Clusters and Noise Points")
plt.show()
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels = dbscan.fit_predict(data)
```

```python
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)

colors = [plt.cm.Spectral(each)
          for each in np.linspace(0, 1, n_clusters)]

for cluster_id, color in zip(range(n_clusters), colors):
    cluster_points = data[labels == cluster_id]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], s=10, color=color)

noise_points = data[labels == -1]
plt.scatter(noise_points[:, 0], noise_points[:, 1], s=10, color='black')

plt.title("DBSCAN Clustering with Anomaly Detection")
plt.show()
```
